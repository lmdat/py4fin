<h2 style="color: blue">1. Matplotlib là gì? Cài đặt</h2>

Matplotlib là một thư viện hỗ trợ người dùng vẽ các biểu đồ (chart) trong Python. Matplotlib có thể vẽ được nhiều dạng chart như: line, bar, scatter, histogram...

Để cài đặt Matplotlib, ta dùng lệnh:

**pip install matplotlib**

Sử dụng:

**import matplotlib.pyplot as plt**

```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2 style="color: blue">2. Vẽ một chart cơ bản</h2>

Thử biểu diễn 2 hàm số sau:

$\Large y=3x+1$

$\Large y=2x^2 + 1$

```python
# y = 3x - 1
x1 = np.linspace(-3, 5, 10)
y1 = (3 * x1) - 1
print("X:", x1)
print("Y:", y1)
```

```python
plt.plot(x1, y1, 'red') 
plt.show()
```

```python
# y = 2x^2 + 1
x2 = np.linspace(-4, 4, 20)
y2 = 2 * (x2 ** 2) + 1
print("X:", x2)
print("Y:", y2)
```

```python
plt.plot(x2, y2, 'b--')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.title('Parabol')
plt.show()
```

<h2 style="color: blue">3. Vẽ nhiều chart trong một hình (figure)</h2>

<h3 style="color: blue">3.1. Dùng hàm subplot</h3>
Dùng hàm subplot để plot từng chart theo vị trí dòng cột

**subplot(rows, cols, plot_number)**
 - rows: số lượng dòng
 - cols: số lượng cột
 - plot_number: số thứ tự của chart

Ví dụ:
- Muốn vẽ 2 chart theo chiều dọc, nghĩa là 2 dòng 1 cột ==> rows=2 và cols=1
- Muốn vẽ 3 chart theo chiều ngang, nghĩa là 1 dòng 3 cột ==> rows=1 và cols=3


```python
# Vẽ 2 chart theo chiều dọc
fig1 = plt.figure(figsize=(5,10))

chart1 = plt.subplot(2, 1, 1)
chart1.plot(x1, y1, 'r') 
chart2 = plt.subplot(2, 1, 2)
chart2.plot(x2, y2, 'b--')

fig1.suptitle('Vẽ nhiều chart trong một hình')
fig1.set_tight_layout(True)
plt.show()
```

```python
import matplotlib.gridspec as gridspec
```

```python
# Vẽ 3 chart theo chiều ngang sử dụng GridSpec
fig2 = plt.figure(figsize=(18,5))
grid = gridspec.GridSpec(1, 3)

chart1 = plt.subplot(grid[0, 0]) #plt.subplot(1, 3, 1)
chart1.set_xlabel('X1')
chart1.set_ylabel('Y1')
chart1.set_title('Line - [0, 0]')

# chart1.plot(x1, y1, color='r', linewidth=2, linestyle='-', marker='o') # linewidth=2 or lw=2 | linestyle='-' or ls='-'
chart1.plot(x1, y1, 'ro-', lw=2)

#--------------------------------------------
chart2 = plt.subplot(grid[0, 1]) #plt.subplot(1, 3, 2)
chart2.set_xlabel('X2')
chart2.set_ylabel('Y2')
chart2.set_title('Parabol 1 - [0, 1]')

chart2.grid(True)
chart2.plot(x2, y2, 'b*--', lw=2)
# chart2.plot(x2, y2, 'b--', lw=3, marker='o')

#--------------------------------------------
chart3 = plt.subplot(grid[0, 2]) #plt.subplot(1, 3, 3)
chart3.set_xlabel('X3')
chart3.set_ylabel('Y3')
chart3.set_title('Parabol 2 - [0, 2]')

chart3.plot(y2, x2, 'g*-')

fig2.suptitle('Vẽ nhiều chart trong một hình')
fig2.set_tight_layout(True)
plt.show()
```

<h3 style="color: blue">3.2. Dùng hàm subplots</h3>

```python
# Vẽ 3 chart theo chiều ngang
fig3, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))
axes[0].set_xlabel('X1')
axes[0].set_ylabel('Y1')
axes[0].set_title('Line')
axes[0].plot(x1, y1, 'r')

#--------------------------------------------
axes[1].grid(True)
axes[1].set_xlabel('X2')
axes[1].set_ylabel('Y2')
axes[1].set_title('Parabol 1')
axes[1].plot(x2, y2, 'b--')

#--------------------------------------------
axes[2].set_xlabel('X3')
axes[2].set_ylabel('Y3')
axes[2].set_title('Parabol 2')
axes[2].plot(y2, x2, 'g*-')

fig3.suptitle('Vẽ nhiều chart trong một hình')
fig3.set_tight_layout(True)
plt.show()
```

```python
# 2x2 chart
fig4, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes[0,0].set_xlabel('X1')
axes[0,0].set_ylabel('Y1')
axes[0,0].set_title('Line - Axes[0,0]')
axes[0,0].plot(x1, y1, 'r')

#--------------------------------------------
axes[0,1].grid(True)
axes[0,1].set_xlabel('X2')
axes[0,1].set_ylabel('Y2')
axes[0,1].set_title('Parabol 1 - Axes[0,1]')
axes[0,1].plot(x2, y2, 'b--')

#--------------------------------------------
axes[1,0].set_xlabel('X3')
axes[1,0].set_ylabel('Y3')
axes[1,0].set_title('Parabol 2 - Axes[1,0]')
axes[1,0].plot(y2, x2, 'g*-')

axes[1,1].set_visible(False)

fig4.suptitle('Vẽ nhiều chart trong một hình')
fig4.set_tight_layout(True)
plt.show()
```

<h3 style="color: blue">3.3. Dùng hàm add_subplot</h3>

```python
fig5 = plt.figure(figsize=(12,8), tight_layout=True)
grid = gridspec.GridSpec(2, 2)

ax1 = fig5.add_subplot(grid[0, :])
ax1.set_xlabel('X1')
ax1.set_ylabel('Y1')
ax1.set_title('Line - Axes[0,:]')
ax1.plot(x1, y1, 'r')

#--------------------------------------------
ax2 = fig5.add_subplot(grid[1, 0])
ax2.grid(True)
ax2.set_xlabel('X2')
ax2.set_ylabel('Y2')
ax2.set_title('Parabol 1 - Axes[1,0]')
ax2.plot(x2, y2, 'b--')

#--------------------------------------------
ax3 = fig5.add_subplot(grid[1, 1])
ax3.set_xlabel('X3')
ax3.set_ylabel('Y3')
ax3.set_title('Parabol 2 - Axes[1,1]')
ax3.plot(y2, x2, 'g*-')

fig5.suptitle('Vẽ nhiều chart trong một hình')
fig5.set_tight_layout(True)
plt.show()
```

<h2 style="color: blue">5. Histogram chart</h2>

Histogram là đồ thị thể hiện sự phân bổ các giá trị của dữ liệu

```python
np.random.seed(2023)
arr = np.random.randint(0, 11, 100)
arr
```

```python
# plt.hist(arr, bins=10)
# plt.hist(arr, bins=10, color='green') # Đổi màu
plt.hist(arr,bins=10, color='green', histtype='step') # Chỉ hiển thị line
plt.xticks(np.arange(0, 11))
plt.show()
```

<h2 style="color: blue">6. Bar chart</h2>

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [800, 1200, 500, 600, 300]
profit = [300, 700, 250, 200, 150]

plt.bar(months, revenue, color='orange')
plt.show()
```

```python
plt.bar(months, revenue, color='orange')
plt.bar(months, profit, color='green')
plt.show()
```

```python
x = np.arange(5) # 0 -> 4
w = 0.4
plt.bar(x - w/2, revenue, width=w, color='orange', label='Revenue')
plt.bar(x + w/2, profit, width=w, color='green', label='Profit')
plt.xticks(x, labels=months)
plt.legend()
plt.show()
```

<h2 style="color: blue">7. Scatter chart</h2>

Scatter là biểu đồ phân tán được sử dụng để hiển thị giá trị và mối quan hệ giữa hai biến định lượng của một tập dữ liệu (biến phụ thuộc và biến độc lập)
- biến phụ thuộc: thường plot trên trục tung
- biến phụ độc lập: thường plot trên trục hoành

Chúng ta sẽ dùng bộ dữ liệu kinh tế lượng của Wooldridge (ông là tác giả của quyển Introductory Econometrics: A Modern Approach) để vẽ scatter chart
- **pip install wooldridge**

```python
import wooldridge as wdg
```

```python
# Xem tất cả bộ dữ liệu
wdg.data()
```

```python
# dùng bộ dữ liệu wage1
wdg.data('wage1', description=True)
```

```python
df = wdg.data('wage1')
df
```

```python
# Xem mối quan hệ giữa biến phụ thuộc: wage và biến độc lập: exper
fig = plt.figure(figsize=(10,6))
plt.scatter(x=df['exper'], y=df['wage'])
plt.xlabel('exper')
plt.ylabel('wage')
plt.show()
```

```python
df2 = wdg.data('hprice1')
df2
```

```python
# Xem mối quan hệ giữa biến phụ thuộc: price và biến độc lập: sqrft
fig2 = plt.figure(figsize=(10,6))
plt.scatter(x=df2['sqrft'], y=df2['price'], color='red')
plt.xlabel('sqrft')
plt.ylabel('price')
plt.show()
```

```python
fig2 = plt.figure(figsize=(10,6))
plt.scatter(x='sqrft', y='price', data=df2, color='r')
plt.xlabel('sqrft')
plt.ylabel('price')
plt.show()
```
