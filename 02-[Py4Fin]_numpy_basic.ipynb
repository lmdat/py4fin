<h3 style="color: blue">2.2. Tạo mảng bằng hàm arange()</h3>
$$
\Large arange(start, stop, step)
$$


**Tổng quát hàm arange():**
- Số phần tử = (stop - start) / step (làm tròn lên: 4.5 --> 5)
- Giá trị phần tử thứ [i + 1] = giá trị phần tử thứ [i] + step
- Giá trị của phần tử cuối cùng luôn nhỏ hơn stop (hay nói cách khác, không nhận giá trị stop)

```python
# hàm arange(start, stop, step)
arrA = np.arange(10) # start = 0, stop = 10, step = 1 (default) |==> số phần tử = (10 - 0) / 1 = 10
print("arrA", arrA, type(arrA))

arrB = np.arange(1, 10) # start = 1, stop = 10, step = 1 |==> số phần tử = (10 - 1) / 1 = 9
print("arrB", arrB)

arrC = np.arange(1, 12, 2) # start = 1, stop = 12, step = 2 |==> số phần tử = (12 - 1) / 2 = 5.5 ~ 6
print("arrC", arrC)
```

<h3 style="color: blue">2.3. Tạo mảng bằng hàm linspace()</h3>
$$
\Large linspace(start, stop, num=50)
$$

**Tổng quát hàm linspace():**
- Số phần tử = num
- Các phần tử có giá trị trong khoản [start; stop] (phần tử đầu tiên có giá trị = start; phần tử cuối cùng có giá trị = stop)
- Các giá trị có khoản cách đều nhau (giá trị sẽ do hàm này tự tính)

```python
# Hàm linspace(start, stop, num)
arr1 = np.linspace(0, 20, 5)
print("arr1", arr1)

arr2 = np.linspace(0, 20, 2)
print("arr2", arr2)

arr3 = np.linspace(0, 20, 3)
print("arr3", arr3)

arr4 = np.linspace(0, 20, 4)
print("arr4", arr4)
```

Gọi hàm seed(num) trước khi gọi các hàm random dùng để cố định cái random state trong thuật toán tạo số random 
- ==> nếu cùng một con số seed thì chạy bất kỳ máy tính nào cũng sẽ cho ra con số ramdom giống nhau

np.random.seed(2023)

#### Hàm random()
$$
\Large random.random(shape)
$$
Hàm này tạo array với các giá trị ngẫu nhiên nằm trong khoảng [0; 1)

```python
#Hàm random()
np.random.seed(2023)
arr13 = np.random.random(3)
print("arr13\n", arr13)

arr14 = np.random.random([2, 4])
print("arr14\n", arr14)
```

#### Hàm rand()
$$
\Large random.rand(d0, d1, ..., dn)
$$
- tạo mảng có các giá trị ngẫu nhiên tuân theo phân phối đồng nhất (Uniform Distribution)
- Cũng giống như random() nhưng thay vì truyền vào một shape(dòng, cột) thì sẽ truyền dòng, cột là từng tham số

```python
# Hàm rand()
np.random.seed(2023)
arr15 = np.random.rand(5) # 1 dòng gồm 5 phần tử
print("arr15\n", arr15)

arr16 = np.random.rand(2, 4) # 2 dòng, mỗi dòng gồm 4 phần tử
print("arr16\n", arr16)
```

#### Hàm normal()
$$
\Large random.normal(loc=0.0, scale=1.0, size=None)
$$
- Tạo mảng các giá trị ngẫu nhiên tuân theo phân phối thông thường (Normal Distribution)

```python
# Hàm normal()
np.random.seed(2023)
arr17 = np.random.normal(size=3)
print("arr17\n", arr17)
```

```python
np.random.seed(2023)
mu = 2 # muy
sigma = 3.2
arr18 = np.random.normal(mu, sigma, size=[2, 4])
print("arr18\n", arr18)
```

#### Hàm randn()
$$
\Large random.randn(d0, d1, ..., dn)
$$
- Tạo mảng các giá trị ngẫu nhiên tuân theo phân phối chuẩn thông thường (Standard Normal Distribution)
- Là trường hợp đặc biệt của hàm normal() với mu=0 và sigma=1

```python
# Hàm randn()
np.random.seed(2023)
arr19 = np.random.randn(3)
print("arr19\n", arr19)
```

```python
np.random.seed(2023)
arr20 = np.random.randn(2, 4)
print("arr20\n", arr20)
```

#### Hàm randint()
$$
\Large random.randint(low, high=None, size=None)
$$
**Tổng quát hàm randint():**

- Các giá trị tạo ra là số nguyên ngẫu nhiên có thể trùng nhau trong khoản [low, high) |==> nhận low và không bao gồm high
- Tạo mảng 1 chiều: size = số nguyên
- Tạo mảng 2 chiều: size = (số dòng, số cột) hoặc [số dòng, số cột]

```python
# mảng một chiều có size = 1
np.random.seed(2023)
arr21 = np.random.randint(1, 7) 
print("arr21\n", arr21)
```

```python
# mảng 1 chiều có size = 10
np.random.seed(2023)
arr22 = np.random.randint(-4, 2, 10) 
print("arr22\n", arr22)
```

```python
# mảng 2 chiều gồm 2 dòng, 5 cột
np.random.seed(2023)
arr23 = np.random.randint(1, 7, (2, 5)) 
print("arr23\n", arr23)
```

#### Hàm choice()
$$
\Large random.choice(a, size=None, replace=True, p=None)
$$
**Tổng quát hàm choice():**
- Nếu a là số nguyên: các giá trị tạo ra sẽ là các số nguyên dương < a (với a > 0)
- Tạo một số nguyên: không truyền giá trị size
- Tạo mảng 1 chiều: size = số nguyên dương
- Tạo mảng 2 chiều: size = (số dòng, số cột) hoặc [số dòng, số cột]
- Các giá trị tạo ra trong mảng có thể trùng nhau
- Nếu a là một tuple hoặc list các số nguyên thì giá trị tạo ra sẽ là một trong các giá trị trong tuple (list) đó


```python
# Tạo một số nguyên: không truyền giá trị size
np.random.seed(2023)
arr24 = np.random.choice(4)
print("arr24\n", arr24)
```

```python
# Tạo mảng 1 chiều: size = số nguyên dương
np.random.seed(2023)
arr25 = np.random.choice(4, 5)
print("arr25\n", arr25)
```

```python
#Tạo mảng 2 chiều: size = (số dòng, số cột) hoặc [số dòng, số cột]
np.random.seed(2023)
arr26 = np.random.choice(4, (2, 5))
print("arr26\n", arr26)
```

```python
# Nếu a là một tuple/list các số nguyên thì giá trị tạo ra sẽ là một trong các giá trị trong tuple (list) đó
np.random.seed(2023)
arr27 = np.random.choice((-1, 3, 6, 8), 10)
print("arr27\n", arr27)
```

<h3 style="color: blue">6.5. Ma trận nghịch đảo</h3>
Cho ma trận vuông A. Nếu tồn tại ma trận vuông $A^{-1}$ sao cho:
$$\Large A.A^{-1} = I$$
Khi đó \\(A^{-1}\\) được gọi là ma trận nghịch đảo của A

```python
np.random.seed(2023)
matrixA = np.random.randint(-2, 10, (3, 3))
print("matrixA\n", matrixA)
np.linalg.inv(matrixA)
```
