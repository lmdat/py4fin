<h2 style="color: blue">1. Seaborn là gì? Cài đặt</h2>

Seaborn là một thư viện phổ biến dùng để vẽ đồ thị trong phân tích data. Nó được build trên nền của Matplotlib, lý do người ta tạo ra Seaborn là để làm đơn giản hóa các thao tác bất tiện khi dùng Matplotlib, đồng thời Seaborn còn cung cấp thêm các theme khá là đẹp.

Để cài đặt Seaborn, ta dùng lệnh:

**pip install seaborn**

Sử dụng:

**import matplotlib.pyplot as plt**

**import seaborn as sns**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sns.set_theme(style='whitegrid', rc={"figure.figsize":(8,6)})
sns.set(style='darkgrid', rc={"figure.figsize":(8, 6)}) # style="darkgrid", "whitegrid", "dark", "white" and "ticks" 
```

<h2 style="color: blue">2. Data set đi theo Seaborn</h2>

Trong Seaborn, người ta cũng cung cấp sẳn một số data set để làm dữ liệu test

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('mpg') # mpg: Miles Per Gallon
df.head(10)
```

<h2 style="color: blue">3. Histogram plot</h2>

```python
sns.histplot(data=df['mpg'])
plt.show()
```

```python
sns.histplot(data=df, x='mpg', kde=True, bins=10)
plt.show()
```

```python
sns.histplot(df['weight'], kde=True, bins=10, color='r')
plt.show()
```

<h2 style="color: blue">4. KDE (Kernel Density Estimate) plot</h2>

```python
sns.kdeplot(data=df['mpg'])
plt.show()
```

```python
sns.kdeplot(x='mpg', data=df, hue='origin')
plt.show()
```

<h2 style="color: blue">5. Distribution plot</h2>

```python
sns.displot(data=df['mpg'])
plt.show()
```

```python
sns.displot(data=df['mpg'], color='g', kind='hist')
plt.show()
```

```python
sns.displot(data=df['mpg'], color='g', kind='kde')
plt.show()
```

```python
sns.displot(data=df, x='mpg', col='origin', color='r', kind='hist', kde=True)
plt.show()
```

<h2 style="color: blue">6. Scatter plot</h2>

```python
# plt.figure(figsize=(12,8))
sns.scatterplot(x=df['horsepower'], y=df['mpg'], color='r')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='horsepower', y='mpg', data=df, hue='horsepower', size='weight', palette='flare') # palette: flare, husl, Spectral, pastel, coolwarm
plt.show()
```

<h2 style="color: blue">7. Jointplot</h2>

Kết hợp một biểu đồ 2 biến với một biểu đồ đơn biến

```python
sns.jointplot(x='weight',y='mpg',data=df, color='r', height=7) # default kind='scatter'
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',data=df, color='r', height=8, hue='origin') # default kind='scatter'
plt.show()
```

```python
# plt.figure(figsize=(12,6))
sns.jointplot(x='weight',y='mpg',data=df, color='r', height=8, kind='hex')
plt.show()
```

```python
sns.jointplot(x='weight',y='mpg',data=df, color='r', height=8, kind='reg') # kind='reg' : regression line
# plt.figure(figsize=(12,6))
plt.show()
```

<h2 style="color: blue">8. Pairplot</h2>

Pairplot sẽ vẽ các mối quan hệ theo cặp trên toàn bộ khung dữ liệu (đối với các cột số) và hỗ trợ đối số màu sắc (đối với các cột phân loại). Nôm na là nó giống như ma trận tương quan (Correlation), thay vì là ma trận số thì nó là ma trận đồ thị

```python
df.head()
```

```python
df1 = df[['mpg','cylinders','weight', 'horsepower']]
df1.head(5)
```

```python
sns.pairplot(data=df1, height=4)
plt.show()
```

```python
df2 = df[['mpg','cylinders','weight', 'horsepower', 'origin']]
df2.head(5)
```

```python
sns.pairplot(data=df2, height=4, hue='origin', palette='husl') # palette: flare, husl, Spectral, pastel, coolwarm
plt.show()
```

<h2 style="color: blue">9. Boxplot</h2>

<h2 style="color: blue">10. Barplot</h2>

```python
plt.figure(figsize=(10,6))
sns.barplot(x='cylinders', y='mpg', data=df)
# sns.barplot(x='cylinders', y='mpg', data=df, facecolor=(0, 0, 0, 0), edgecolor='.5', capsize=0.4)
plt.show()
```

```python
df[['mpg', 'cylinders']].groupby('cylinders').mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='cylinders', y='mpg', data=df, hue='origin')
plt.show()
```

<h2 style="color: blue">11. Countplot</h2>

```python
plt.figure(figsize=(10,6))
sns.countplot(x='origin',data=df)
plt.show()
```

```python
df['origin'].value_counts()
```

<h2 style="color: blue">12. Heatmap</h2>

```python
df.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True) # cmap = 'flare' | 'coolwarm' | 'Spectral'
plt.show()
```

<h2 style="color: blue">13. Vẽ nhiều chart trong một figure</h2>

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 12))

sns.histplot(df['weight'], kde=True, bins=10, color='r', ax=axes[0,0])
sns.boxplot(x='cylinders', y='mpg', data=df, ax=axes[0,1])
sns.scatterplot(x=df['horsepower'], y=df['mpg'], color='r', ax=axes[1,0])
sns.barplot(x='cylinders', y='mpg', data=df, ax=axes[1,1])

fig.suptitle('Vẽ nhiều chart trong một hình với Seaborn')
plt.show()
```
